```python
year=np.arange(1980,1991)
enrollment=np.array([10,80,50,65,33,77,99,74,66,44,55])
plt.plot(year,enrollment, color='red', marker="o",linestyle='--',linewidth='2')
plt.xlabel('years');
plt.ylabel('enrolment')
plt.grid(True)
plt.show()
```

```python
Categories=np.array(["Bajaj","Tuition fee","shopping", "makeup"])
Expenses=np.array([7,60,30,20])
colors=["blue","yellow","green","red"]

# visualisation
plt.bar(Categories,Expenses, color=colors)
plt.title('Monthly Expenses')
plt.xlabel('Catagories')
plt.ylabel('Expenses')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
expenses_pie=['rent','groceries','utilities','entertainment']
cost_pie=[100,200,100,150]

plt.pie(cost_pie, labels=expenses_pie, autopct='%1.1f%%')
plt.title("monthly expenses distrobution. ")
plt.show()
```

# Plot a line chart showing the number of tourists visiting a country from 2010 to 2020. Use the following data:

```python
import numpy as np
import matplotlib.pyplot as plt
years=np.arange(2010,2020)
tourist=np.array([500,550,600,700,750,800,850,950,1000,1100])

plt.plot(years,tourist, marker='o',color="red")
plt.grid(True)
plt.show()
```
